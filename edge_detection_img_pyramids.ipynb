```python
#Load the provided image “texture.jpeg” and convert it to grayscale
import cv2
import numpy as np
im = cv2.imread('/Users/harshikajha/Downloads/EE379V_Spring2022_HW1/texture.jpeg', 0)
cv2.imshow('', im)
cv2.waitKey(1)
```

```python
#Write code for a Gaussian and Laplacian pyramid of 5 levels (use for loops)
```

```python
#Gaussian Pyramid
layer=im.copy()
gaussian_pyramid=[layer]
for i in range(6):
    layer=cv2.pyrDown(layer)
    gaussian_pyramid.append(layer)
    cv2.imshow(str(i+1),gaussian_pyramid[i+1])
    cv2.waitKey(1)
```

```python
#Laplacian Pyramid
for i in range(5,0,-1):
    size=(gaussian_pyramid[i-1].shape[1],gaussian_pyramid[i-1].shape[0])
    ext=cv2.pyrUp(gaussian_pyramid[i],dstsize=size)
    laplacian=cv2.absdiff(gaussian_pyramid[i-1],ext)
    cv2.imshow(str(i+1),laplacian)
    cv2.waitKey(1)
```

```python
import numpy
from math import sqrt
import matplotlib.pyplot as plt 
import skimage
from skimage import color, data, filters
import scipy
from scipy.ndimage import gaussian_filter

from skimage import io

def gradientMagnitude(im,sigma):
    
    im = scipy.ndimage.gaussian_filter(im, sigma)
    plt.imshow(im)
    plt.show()
    
    ex2_height, ex2_width = im.shape
    ex2_v = skimage.filters.sobel_v(im)
    ex2_h = skimage.filters.sobel_h(im)
    
    mag = (ex2_h ** 2 + ex2_v ** 2) ** (1/2)
    theta = np.arctan2(ex2_h,ex2_v)
    
    return mag, theta

gradientMagnitude(im,0)
```

```python
import cv2 
import os 
import numpy as np
from  skimage.feature import canny
def edgeOrientedFilters(im):
    [mag, theta] = gradientMagnitude(im,3)
    suppressed = np.zeros(mag.shape)
    h, w = mag.shape
    angles = np.rad2deg(theta)
    angles[angles < 0] += 180
    for i in range(h - 1):
        for j in range(w - 1):
            p, k = 255, 255
            theta = angles[i,j]
            if (theta >= 0 and theta < 22.5) or (theta >= 157.5 and theta <= 180):
                # angle 0
                p = mag[i, j-1]
                k = mag[i, j+1]
            elif (theta >= 22.5 and theta < 67.5):
                # angle 45
                p = mag[i+1, j-1]
                k = mag[i-1, j+1]
            elif (theta >= 67.5 and theta < 112.5):
                # angle 90
                p = mag[i+1, j]
                k = mag[i-1, j]
            elif (theta >= 112.5 and theta < 157.5):
                # angle 135
                p = mag[i-1, j-1]
                k = mag[i+1, j+1]
            if mag[i,j] >= p and mag[i,j] >= k:
                suppressed[i,j] = mag[i,j]
    return suppressed
    
plt.imshow(edgeOrientedFilters(im))
```

```python
def gradientMagnitude2(im):
    prev = 0
    final_thet = 0
    for i in range(4):
        ddepth = cv2.CV_32F
        sobel_x = cv2.Sobel(im, ddepth, 1, 0)
        sobel_y = cv2.Sobel(im, ddepth, 0, 1)
        dxabs = cv2.convertScaleAbs(sobel_x)
        dyabs = cv2.convertScaleAbs(sobel_y)
        mag = cv2.addWeighted(dxabs, 0.5, dyabs, 0.5, 0)
        theta = np.arctan2(dyabs,dxabs)
        
    return mag,theta
    
gradientMagnitude2(im)
```
